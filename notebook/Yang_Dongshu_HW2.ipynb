```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import zscore
from itertools import combinations

df = pd.read_excel("../data/Folds5x2_pp.xlsx")

# 1. (b) i.
rows, cols = df.shape
print(f"Rows: {rows}, Columns: {cols}")
#For rows, each row represents a set of inputs. 
#For columns, the first four columns are predictor variables, and the last column is the outcome variable.
print('*' * 100)

# 1. (b) ii.
sns.pairplot(df, corner = True)
plt.show()
# description: from the scatter plot, 
# it seems that the negative correlation between AT and V with PE is relatively significant,
# while the influence of AP and RH may be relatively small.
print('*' * 100)

# 1. (b) iii. 
summary = df.describe().T
summary["Range"] = summary["max"] - summary["min"]
summary["IQR"] = summary["75%"] - summary["25%"]
print(summary)
print('*' * 100)

# 1. (c) Here I use the statsmodel suggested by professor
predictors = [col for col in df.columns if col != "PE"]
uni_coef = {} # store data for 1.(e)

for predictor in predictors:
    X = sm.add_constant(df[predictor])
    y = df['PE']
    model = sm.OLS(y, X).fit()
    print(f"\nLinear Regression for {predictor}:")
    print(model.summary())
    uni_coef[predictor] = model.params.iloc[1] # store data for 1.(e)
    # create plots for assertion
    plt.figure(figsize=(6,4))
    sns.regplot(x=df[predictor], y=y, scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
    plt.xlabel(predictor)
    plt.ylabel("PE")
    plt.title(f"Regression of PE on {predictor}")
    plt.show()
    # detect outliers, here I use Z score
    df["z_score"] = zscore(df[predictor])
    outliers = df[df["z_score"].abs() > 3]
    if not outliers.empty:
        print(f"Outliers detected for {predictor}:\n", outliers)
# description: The data indicates that 
# AT and V significantly negatively predict PE,
# while AP and RH significantly positively predict PE (p < 0.001)
print('*' * 100)

# 1. (d) multiple regression model
X = df.drop(columns=["PE"]) 
y = df["PE"]
X = sm.add_constant(X)  
model = sm.OLS(y, X).fit()
print(model.summary())
# Still, P < 0.001 for every predictor, so we reject the null hypothesis for all predictors
print('*' * 100)

# 1. (e) multi vs uni 
multi_coef = model.params.drop(["const",'z_score'])
plt.figure(figsize=(7,7))
plt.scatter(uni_coef.values(), multi_coef.values, color='blue', alpha=0.7, edgecolors='k')

plt.xlabel("Uni Regression Coef")
plt.ylabel("Multiple Regression Coef")
plt.title("Comparison of Uni and Multi")

plt.plot([-3, 6], [-3, 6], linestyle="--", color="gray") 
# If the univariate and multiple regression coefficients are equal, the data points should lie on the gray dashed line y=x.

for predictor in uni_coef.keys():
    x = uni_coef[predictor]
    y = multi_coef[predictor]
    plt.text(x, y, f"{predictor}\n({x:.2f}, {y:.2f})", fontsize=10, ha='right', color='red')

plt.show()
#print(uni_coef)
#print(multi_coef)
print('*' * 100)

# 1. (f) nonlinear regression
print("\nTesting for Nonlinear Association")
for predictor in predictors:
    df[f"{predictor}^2"] = df[predictor] ** 2
    df[f"{predictor}^3"] = df[predictor] ** 3
    X_poly = sm.add_constant(df[[predictor, f"{predictor}^2", f"{predictor}^3"]])
    y = df['PE']
    model_poly = sm.OLS(y, X_poly).fit()
    print(f"\nPolynomial Regression for {predictor} :")
    print(model_poly.summary())
# The results show that except for the quadratic term of V is not significant (P=0.768)， 
# and the cubic term of V is statistically significant (P=0.014 < 0.05),
# all other results are highly significant (P<0.001)
print('*' * 100)

# 1. (g) interactions analysis
print("\nTesting for Interaction Effects")
interaction_terms = []
for var1, var2 in combinations(predictors, 2):
    interaction_name = f"{var1}*{var2}"
    df[interaction_name] = df[var1] * df[var2]
    interaction_terms.append(interaction_name)

X_interactions = sm.add_constant(df[predictors + interaction_terms])
y = df['PE']
model_interactions = sm.OLS(y, X_interactions).fit()
print(model_interactions.summary())
# The results show that except for the interactions AT * AP (p = 0.452) and V * RH (p = 0.086),
# all other predictor pairs exhibit statistically significant interaction effects.
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import numpy as np
from itertools import combinations
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import mean_squared_error as mse
from sklearn.neighbors import KNeighborsRegressor as KNN
from sklearn.preprocessing import StandardScaler as SS

df = pd.read_excel("../data/Folds5x2_pp.xlsx")
train_df, test_df = tts(df, test_size=0.3, random_state=42)

# 1. (h)
X_train = train_df.drop(columns=["PE"])
X_train = sm.add_constant(X_train)
y_train = train_df["PE"]
baseline_model = sm.OLS(y_train, X_train).fit()
#print("Baseline Model Summary:")
#print(baseline_model.summary())

predictors = [col for col in df.columns if col != "PE"]
interaction_terms = []
nonlinear_terms = []
for var1, var2 in combinations(predictors, 2):
    interaction_name = f"{var1}*{var2}"
    train_df[interaction_name] = train_df[var1] * train_df[var2]
    test_df[interaction_name] = test_df[var1] * test_df[var2]
    interaction_terms.append(interaction_name)

for predictor in predictors:
    square_term = f"{predictor}^2"
    train_df[square_term] = train_df[predictor] ** 2
    test_df[square_term] = test_df[predictor] ** 2
    nonlinear_terms.append(square_term)

X_train_final = sm.add_constant(train_df[predictors + interaction_terms + nonlinear_terms])
final_model = sm.OLS(y_train, X_train_final).fit()
print("Final Model:")
print(final_model.summary())
print('*' * 100)

significant_vars = final_model.pvalues[final_model.pvalues < 0.05].index.tolist()
if "const" in significant_vars:
    significant_vars.remove("const")
X_train_filtered = sm.add_constant(train_df[significant_vars])
filtered_model = sm.OLS(y_train, X_train_filtered).fit()
print("Filtered Model Summary:")
print(filtered_model.summary())
print('*' * 100)

X_test = sm.add_constant(test_df[predictors])
y_test = test_df["PE"]
base_mse_train = mse(y_train, baseline_model.predict(X_train))
base_mse_test = mse(y_test, baseline_model.predict(X_test))

X_test_filtered = sm.add_constant(test_df[significant_vars])
filtered_mse_train = mse(y_train, filtered_model.predict(X_train_filtered))
filtered_mse_test = mse(y_test, filtered_model.predict(X_test_filtered))

print(f"Baseline Model Train MSE: {base_mse_train:.2f}, Test MSE: {base_mse_test:.2f}")
print(f"Filtered Model Train MSE: {filtered_mse_train:.2f}, Test MSE: {filtered_mse_test:.2f}")
print('*' * 100)

# 1. (i)

X_train_knn = train_df.drop(columns=["PE"])
y_train_knn = train_df["PE"]
X_test_knn = test_df.drop(columns=["PE"])
y_test_knn = test_df["PE"]

scaler = SS()
X_train_knn_scaled = scaler.fit_transform(X_train)
X_test_knn_scaled = scaler.transform(X_test)

train_errors = []
test_errors = []

for k in range(1, 101):
    knn = KNN(n_neighbors=k)
    knn.fit(X_train_knn_scaled, y_train_knn)
    train_errors.append(mse(y_train_knn, knn.predict(X_train_knn_scaled)))
    test_errors.append(mse(y_test, knn.predict(X_test_knn_scaled)))

plt.figure(figsize=(8,6))
plt.plot(1 / np.array(range(1, 101)), train_errors, label='Train MSE', marker='o')
plt.plot(1 / np.array(range(1, 101)), test_errors, label='Test MSE', marker='s')
plt.xlabel("1/k")
plt.ylabel("MSE")
plt.title("KNN Regression: MSE vs 1/k")
plt.legend()
plt.show()

best_k = np.argmin(test_errors) + 1
print(f"Best k value: {best_k}, Test MSE: {min(test_errors):.2f}")
```
